```python
import pandas as pd
import re
```

```python
path='三通局5月过坝计划汇总.xlsx'
# path='0519三通局.xlsx'
```

```python
def get_square_rate(row):
#     print(row)
#     for k,v in enumerate(row):
#         print(k,v)
    length=len(row)//2
    print('length',length)
    s=0
    for i in range(length):
        ind=2*i+1
        print('row_ind',row[ind],type(row[ind]))
        if pd.notnull(row[ind])  :
            l_w=re.split('[,，]',str(row[ind]))
            print('l_w',l_w)
            l=l_w[0]
            w=l_w[1]
            s=s+float(l)*float(w)
    rate=s/(34*280)
    print(s,rate)
    return rate
```

```python

with pd.ExcelWriter(s_path) as writer:
    for sheet_name,e_df in dd.items():
    #     print(sheet_name,e_df)
#           if sheet_name in ['【完成】0501','【完成】0509']:
#               pass
              print('sheet_name',sheet_name)
              aa=e_df.iloc[2:,3:].dropna(how='all',axis=1)
    #           aa=e_df.dropna(how='all',axis=1)
              aa.columns=range(aa.shape[1])
             
              aa['use_rate']=aa.apply(lambda x:get_square_rate(x),axis=1)

              aa.to_excel(writer, sheet_name=sheet_name,index=False)
```
